**Exercise 5**. Run the code cell below to display a *choropleth* showing election results for the 2013 Montreal mayoral election.

(a) What would be a reason why you might expect two candidates to get their highest vote totals in the same part of town? What would be a reason you might expect two candidates to get their highest vote totals in *different* parts of town?

(b) Change `"Bergeron"` to `"Coderre"`. Did the two candidates get their highest vote totals in a similar part of town or quite different?

```python
import plotly.express as px

df = px.data.election()
geojson = px.data.election_geojson()

fig = px.choropleth(df, geojson=geojson, color="Coderre",
              locations="district", featureidkey="properties.district",
              projection="mercator")
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
